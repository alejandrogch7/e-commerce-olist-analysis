2. Now we can build the main dataframe to work with 

- Starts with merging orders with customers

```python
df = orders.merge(customers, on='customer_id', how='left') 
```

- Addign payments

```python
df = df.merge(payments, on='order_id', how='left')
```

- Adding order items

```python
df = df.merge(order_items, on='order_id', how='left')
```

- Adding Products

```python
df = df.merge(products, on='product_id', how='left')
```

- Adding Reviews

```python
df = df.merge(reviews, on='order_id', how='left')
```

3. Exploring nulls

```python
df.isnull().sum().sort_values(ascending=False)
```

- We need to get rid of those null values that we can not evaluate 

```python
df = df.dropna(
    subset=[
        'product_id',
        'price',
        'freight_value'
    ]
)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

- Also we noticed that the reviews score are too low, so we are going to fill results with the median

```python
df['review_score'] = (
    df['review_score']
    .fillna(df['review_score'].median())
)
```

- And now we are going to filter for orders that are already delivered

```python
df = df[df['order_status'] == 'delivered']
```

4. Convert the dates to use them

```python
df.columns
```

```python
date_columns = [
    'order_purchase_timestamp',
    'order_approved_at',
    'shipping_limit_date',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date',
    'review_creation_date',
    'review_answer_timestamp'
]

for col in date_columns:
    df[col] = pd.to_datetime(df[col])
```

```python
df.info()
```

5. Creating Derived Columns

```python
df['total_value'] = (
    df['price'] + df['freight_value']
)

df['delivery_delay'] = (
    df['order_delivered_customer_date']
    - df['order_estimated_delivery_date']
).dt.days

df['purchase_year'] = (
    df['order_purchase_timestamp'].dt.year
)

df['purchase_month'] = (
    df['order_purchase_timestamp'].dt.month
)

df['purchase_month_name'] = (
    df['order_purchase_timestamp']
    .dt.month_name()
)
```

6. Obtaining KPIs

Total Revenue
Total Orders
Average Order Value (grouped by order ID)
Average Review Score
Delayed Percentage
Most Used Payment Method
Top Categories
Revenue by State
Monthly Revenue

```python
total_revenue = df['total_value'].sum()

total_orders = df['order_id'].nunique()

average_order_value = (
    df.groupby('order_id')['total_value']
    .sum()
    .mean()
)

average_review = df['review_score'].mean()

df['is_delayed'] = df['delivery_delay'] > 0
delayed_percentage = (
    df['is_delayed'].mean() * 100
)

most_used_payment = df['payment_type'].value_counts().index[0]

top_categories = (
    df.groupby('product_category_name')
    ['total_value']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

revenue_by_state = (
    df.groupby('customer_state')
    ['total_value']
    .sum()
    .sort_values(ascending=False)
)
monthly_revenue = (
    df.groupby('purchase_month_name')
    ['total_value']
    .sum()
)

## Showing results
print(f"Total Revenue: {total_revenue}")
print(f"Total Orders: {total_orders}")
print(f"Average Order Value: {average_order_value}")
print(f"Average Review Score: {average_review}")
print(f"Delayed Orders Percentage: {delayed_percentage:.2f}%")
print(f"Most Used Payment Method: {most_used_payment}")
print(f"Top Categories:\n{top_categories}")
print(f"Revenue by State:\n{revenue_by_state}")
print(f"Monthly Revenue:\n{monthly_revenue}")
```

7. Obtaining valuable insights

- Which categories are generating the most revenue?

```python
top_categories = (
    df.groupby('product_category_name')
    ['total_value']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_categories
```

- Which states generate the most sales?

```python
revenue_by_state = (
    df.groupby('customer_state')
    ['total_value']
    .sum()
    .sort_values(ascending=False)
)

revenue_by_state.head(10)
```

- Do late deliveries affect reviews?

```python
df.groupby('is_delayed')['review_score'].mean()
```

- Which categories have best/worst reviews?

```python
review_counts = (
    df.groupby('product_category_name')
    ['review_score']
    .count()
)

review_means = (
    df.groupby('product_category_name')
    ['review_score']
    .mean()
)

category_reviews = pd.DataFrame({
    'review_count': review_counts,
    'average_review': review_means
})

category_reviews = (
    category_reviews[
        category_reviews['review_count'] >= 100
    ]
    .sort_values(
        by='average_review',
        ascending=False
    )
)

category_reviews.head(10)
```
